# Two options for subplots


```python
# subplot #1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                            ncols = 2,
                                            figsize = (10, 5))

# plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# subplot #2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10, 5))
#plot to each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

# Use subplot from OO

* When plotting something quick and dirty use pyplot method
* When plotting something more advance use OO Method

```python
# subset creation
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
# pyplot method
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target');
```

```python
# OO Method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax)

#ax.set_xlim([45, 100])
```

```python
# use OO Method from scratch

fig, ax = plt.subplots(figsize=(10, 6))

# plot data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

# customize
ax.set(title="Heart Disease and cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol");
ax.legend(*scatter.legend_elements(), title="Target")

#add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
over_50.head()
```

```python
# add subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10, 10),
                              sharex=True)

# add Data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# customize ax0
ax0.set(title="Heart disease and Cholesterol Levels",
       ylabel="chol")

# add legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")
# add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");

# add ax1 with thalach 
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])


# add legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")
# add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");
# customize ax1
ax1.set(title="Heartdisease and Max Heart Rate",
       xlabel="age",
       ylabel="max heart rate");

# add title to figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight='bold');
```

```python
# set style
plt.style.use('seaborn-whitegrid')

# use OO Method from scratch

fig, ax = plt.subplots(figsize=(10, 6))

# plot data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter") # this changes the col. scheme

# customize
ax.set(title="Heart Disease and cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol");
ax.legend(*scatter.legend_elements(), title="Target")

#add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");

```

```python
# customize y and x axis limitation

# add subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10, 10),
                              sharex=True)

# add Data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter")

# customize ax0
ax0.set(title="Heart disease and Cholesterol Levels",
       ylabel="chol")
# change x axes limits
ax0.set_xlim([50, 80])

# add legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")
# add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");

# add ax1 with thalach 
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"],
                     cmap="winter")


# add legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")
# add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");
# customize ax1
ax1.set(title="Heartdisease and Max Heart Rate",
       xlabel="age",
       ylabel="max heart rate");
# change x axes limits
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

# add title to figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight='bold');
```
